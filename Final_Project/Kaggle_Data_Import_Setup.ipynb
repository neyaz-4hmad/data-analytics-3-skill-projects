### Step 3: Authenticate with Kaggle

> Before downloading datasets or competition files, authenticate your Kaggle account.

> There are three common ways to configure the Kaggle API Token.

> Choose **one** of the following methods based on your workflow.

> - **Method 1:** Secure Input using `getpass()`
> - **Method 2:** Environment Variable using `os.environ`
> - **Method 3:** Permanent Configuration using `access_token`
> - **Method 4:** Environment Variables using a `.env` file

> **Important**

> - Never expose your API Token in notebooks, source code, or GitHub repositories.

#### Method 1: Secure Input using `getpass()`

> The `getpass` module securely accepts your API Token without displaying it on the screen.

#### Advantages

> - Hides the API Token while typing.
> - Suitable for learning and demonstrations.
> - Prevents accidental exposure of credentials.

````python
# Import required modules
import os
from getpass import getpass

# Prompt the user to enter the Kaggle API Token securely
api_token = getpass("Enter your Kaggle API Token: ")

# Store the API Token as an environment variable
os.environ["KAGGLE_API_TOKEN"] = api_token

# Display a confirmation message
print("Kaggle API Token configured successfully.")
````

#### Method 2: Configure the API Token using an Environment Variable

> This method stores the API Token only for the current Python session.

#### Advantages

> - No additional files are required.
> - Suitable for testing and temporary sessions.

#### Limitations

> - The API Token must be configured again after restarting the Jupyter kernel.
> - Avoid hardcoding your actual API Token if the notebook will be shared or uploaded to GitHub.

#### Method 3: Configure the API Token using a `.env` File

> Create a file named:

```text
.env
```

> Add the following line.

```text
KAGGLE_API_TOKEN=YOUR_KAGGLE_API_TOKEN
```

> The `.env` file should never be uploaded to GitHub.

> Add it to your `.gitignore` file.

````python
# Import required modules
import os
from dotenv import load_dotenv

# Load environment variables from the .env file
load_dotenv()

# Read the Kaggle API Token
api_token = os.getenv('KAGGLE_API_TOKEN')

# Configure the environment variable
os.environ['KAGGLE_API_TOKEN'] = api_token

# Display a confirmation message
print('Kaggle API Token loaded successfully.')
````
